```python
import ast
import json
```

```python
with open('XL_gossip_history.json', 'rb') as f:
    gossip = json.load(f)
```

```python
x=ast.literal_eval(gossip)
```

```python
train_loss_gossip=x['loss']
valid_loss_gossip=x['val_loss']
```

```python
train_loss_gossip[0:10]
```

```python
valid_loss_gossip[0:10]
```

```python
with open('XL_poli_history.json', 'rb') as f:
    pol = json.load(f)
```

```python
y=ast.literal_eval(pol)
```

```python
train_loss_poli=y['loss']
valid_loss_poli=y['val_loss']
```

```python
train_loss_poli[0:10]
```

```python
valid_loss_poli[0:10]
```

```python
import matplotlib.pyplot as plt
plt.plot(train_loss_gossip)
plt.plot(valid_loss_gossip)
plt.plot(train_loss_poli)
plt.plot(valid_loss_poli)
plt.ylim(0,100)
plt.xlim(0,80)
plt.title('Model loss')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend(['Train_Loss_Gossip', 'Test_Loss_Gossip','Train_Loss_Politifact','Test_Loss_Politifact'], loc='upper right')
plt.show()
```
